```python
def f(z):
    return z**3 -1

def der_f(z):
    return 3*z**2

roots = [1, (-1/2 + 1j*(3**(1/2))/2), (-1/2 - 1j*(3**(1/2))/2)]
colors = [
    [255, 0, 0],  # Red
    [0, 255, 0],  # Green
    [0, 255, 255],  # Blue
]

plot_colored_damped_traub(f, der_f, (-2,2), (-2,2), 1024,1024,roots, colors, delta = 0)
```

```python
def f(z):
    return z*(z-1j)*(z-1)

def der_f(z):
    return (z-1j)*(z-1) + z*(z-1j) + z*(z-1)

roots = [1,1j,0]
colors = [
    [255, 0, 0],  # Red
    [0, 255, 0],  # Green
    [0, 255, 255],  # Blue
]

plot_colored_damped_traub(f, der_f, (-2,2), (-2,2), 1024,1024,roots, colors, delta = 0)
```

```python
def f(z):
    return z**4 -1

def der_f(z):
    return 4*z**3

roots = [1,-1,1j,-1j]
colors = [
    [255, 0, 0],  # Red
    [0, 255, 0],  # Green
    [0, 255, 255],  # Blue
    [255, 255, 0],  # Yellow
]

plot_colored_damped_traub(f, der_f, (-3,3), (-3,3), 1024,1024,roots, colors, delta = 0)
```

```python
def f(z):
    return z*(z-0.2)*(z-(0.5+0.5j))*(z+(0.5+0.5j))*(z-1j)

def der_f(z):
    return 5*(-(0.16+0.8j)*z**3 -(0.2-0.04j)*z + z**4 - (0.18j)*z**2 + 0.02)

roots = [0, 0.2, 0.5+0.5j, -0.5-0.5j, 1j]
colors = [
    [255, 0, 0],  # Red
    [0, 255, 0],  # Green
    [0, 255, 255],  # Blue
    [255, 255, 0],  # Yellow
    [255, 0, 255],  # Purple
]

plot_colored_damped_traub(f, der_f, (-3,3), (-3,3), 1024,1024,roots, colors, delta = 0)
```

```python
def f(z):
    return (z**2 + 0.25)*(z-0.439)

def der_f(z):
    return 2*z*(z-0.439) + (z**2 + 0.25)
roots = [0.439, (-0.25)**(1/2), -(-0.25)**(1/2)]
colors = [
    [255, 0, 0],  # Red
    [0, 255, 0],  # Green
    [0, 255, 255],  # Blue
]

plot_colored_damped_traub(f, der_f, (-0.8,0.8), (-0.8,0.8), 1024,1024,roots, colors, delta = 0)
```
